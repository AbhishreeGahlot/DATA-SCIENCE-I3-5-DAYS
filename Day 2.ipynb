```python
import mysql.connector as ms

try:
    # Try to establish a connection to the database
    conn = ms.connect(
        host="localhost",
        user="root",
        password="",  # Replace with your actual MySQL root password
        database="piet"
    )
    print("Connection successful")

    # If the connection was successful, proceed with executing queries
    mycursor = conn.cursor()
    query = "SELECT * FROM student_data"
    mycursor.execute(query)

    data = mycursor.fetchall()
    for record in data:
        print(record)

except ms.Error as e:
    # Print the error message if connection fails
    print("Error:", e)

finally:
    # Ensure that the connection is closed if it was established
    if 'conn' in locals() and conn.is_connected():
        mycursor.close()
        conn.close()
        print("Connection closed")
```
